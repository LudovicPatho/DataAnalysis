# Visualizing Geographical Data using Basemap
Documentation on Basemap available [here](http://matplotlib.org/basemap/api/basemap_api.html#mpl_toolkits.basemap.Basemap).

## Importing the data

```python
import pandas as pd
airlines = pd.read_csv('airlines.csv')
airports = pd.read_csv('airports.csv')
routes = pd.read_csv('routes.csv')
```

## Airport map

We need to convert the longitude/latitude coordinates available in the airport dataset into Cartesian coordinates.
We will use the Mercator projection for this.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180)
```

```python
x, y = m(airports.longitude.tolist(), airports.latitude.tolist())
```

We plot the result using a scatter plot :

```python
fig, ax = plt.subplots(figsize=(15,20))
plt.title('Scaled Up Earth With Coastlines')
m.scatter(x, y, s=1)
m.drawcoastlines(color=(180/255, 180/255, 180/255))
plt.show()
```

## Routes map
We now plot the airline routes as great circles using `basemap.drawgreatcircle`:

```python
fig, ax = plt.subplots(figsize=(15,20))
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80, llcrnrlon=-180, urcrnrlon=180)
m.drawcoastlines()

def create_great_circles(df):
    for index, row in df.iterrows():
        if abs(row.end_lat - row.start_lat) < 180 and abs(row.end_lon - row.start_lon) < 180:
            m.drawgreatcircle(row.start_lon, row.start_lat, row.end_lon, row.end_lat)

geo_routes = pd.read_csv('geo_routes.csv')            
dfw = geo_routes[geo_routes.source == 'DFW']
create_great_circles(dfw)
plt.show()
```
